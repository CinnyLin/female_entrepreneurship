# kmeans clustering
clustering the majority class before classification so that the dataset is more balanced

```python
# data
import pandas as pd
import numpy as np

# data visualization
import matplotlib.pyplot as plt

# machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# evaluation
from sklearn import metrics
from collections import Counter
```

## data

```python
df = pd.read_csv('../data/combined_feng.csv', index_col=0)
df.shape
```

```python
# get data
X = df.drop(columns='female_led')
y = df['female_led']
```

```python
y.value_counts()
```

```python
# only take X that belongs to the majority class
X = X[y==0]
X.shape
```

### data preprocessing

kmeans clustering result is prone to the size of the numbers so it is important to scale the data before fitting in the model.

```python
scaling=StandardScaler()
X_scaled = scaling.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled)
```

## machine learning

```python
# test out multiple k-values
inertias = []
K = range(1, 10)
 
for k in K:
    mdl = KMeans(n_clusters=k).fit(X)
    inertias.append(mdl.inertia_)
```

```python
# elbow plot
plt.plot(K, inertias, '.-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
# test out multiple k-values
inertias_scaled = []
K = range(1, 10)
 
for k in K:
    mdl = KMeans(n_clusters=k).fit(X_scaled)
    inertias_scaled.append(mdl.inertia_)
```

```python
# elbow plot (X_scaled)
plt.plot(K, inertias_scaled, '.-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
# test out multiple k-values
inertias_scaled = []
K = range(1, 20)
 
for k in K:
    mdl = KMeans(n_clusters=k).fit(X_scaled)
    inertias_scaled.append(mdl.inertia_)
```

```python
# elbow plot (X_scaled)
plt.plot(K, inertias_scaled, '.-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

To determine the optimal number of clusters, we have to select the value of k at the “elbow” ie the point after which the inertia start decreasing in a linear fashion. Thus for the given data, the optimal number of clusters for the data is 3.

```python
k=3
mdl = KMeans(n_clusters=k).fit(X)
Counter(mdl.labels_)
```

```python
# X['kmeans']=mdl.labels_
```

```python
k=3
mdl = KMeans(n_clusters=k).fit(X_scaled)
Counter(mdl.labels_)
```

```python
X['kmeans'].value_counts()
```

Even though `k=3` is the optimal based on elbow plot, the classification is still extremely imbalanced.

### plot cluster features

```python
def scat(x_col='Trend Score (7 Days)', 
         y_col='last_funding_amount_usd', 
         c_col='kmeans'):
    plt.figure(figsize=(10,7))
    plt.scatter(x = X[x_col], y = X[y_col], c = X[c_col])
    plt.xlabel(x_col)
    plt.ylabel(y_col) 
```

```python
scat(x_col, y_col)
```

```python
scat(x_col, y_col)
```

### evaluate classification distribution

```python
def class_dist(k):
    mdl = KMeans(n_clusters=k).fit(X)
    return Counter(mdl.labels_)
```

```python
class_dist(3)
```

```python
class_dist(3)
```

```python
# test out multiple k-values
distribution = {}
K = range(1, 10)
 
for k in K:
    distribution[k] = class_dist(k)
```

```python
dist_df = pd.DataFrame.from_dict(distribution)
```

```python
dist_df = dist_df.T
dist_df.reset_index(inplace=True)
dist_df
```

```python
dist_df.plot(x='index', kind='bar', stacked=True)
```

Even when the number of clusters increase, the imbalanced situation persists.

```python
k=10
c_col = f'kmeans{k}'
mdl = KMeans(n_clusters=k).fit(X)
X[c_col]=mdl.labels_
```

```python
x_col = 'Trend Score (30 Days)'
scat(x_col, y_col, c_col)
```
