# bar plot by region/gender/industry

```python
import pandas as pd
```

```python
regions = ['US', 'EU', 'Europe', 'China']
top = [32, 98]
by_industry_groups = ['industries', 'industry_groups']
```

### plot function

```python
def bar_plot(n):
    
    for region in regions:
        for big in by_industry_groups:
            
            # read data
            if n!=1000:
                for n in top:
                    df = pd.read_csv(f'../data/crunchbase-aggregated/{region}-{n}-{big}.csv')
            
            elif n==1000:
                df = pd.read_csv(f'../data/crunchbase-aggregated/{region}-{big}.csv')
                df = df[['industry', '#female_co', '#not_female_co']]
            
            # plot
            df.set_index('industry')\
              .plot.bar(figsize=(15,6),
                        subplots=True,
                        title=f'{region}-{big}');
```

## top 32 / 98

```python
bar_plot(32)
```

```python
bar_plot(98)
```

## top 1000

```python
bar_plot(1000)
```
